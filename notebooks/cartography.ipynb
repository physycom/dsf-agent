# Downloading Bologna's Catography

Here we just download the cartography, edges and nodes file.

> NOTE: download libtbb-dev with `sudo apt install libtbb-dev`.

```python
from dsf import get_cartography

gdf_edges, gdf_nodes = get_cartography(place_name="Bologna", return_type="gdfs")

# save as csv
gdf_edges.to_csv("../input/bologna_edges.csv", index=False, sep=";")
gdf_nodes.to_csv("../input/bologna_nodes.csv", index=False, sep=";")
# save as geojson
gdf_edges.to_file("../input/bologna_edges.geojson", driver="GeoJSON")
gdf_nodes.to_file("../input/bologna_nodes.geojson", driver="GeoJSON")
```
