# Road Network

Road network is not serializable w/ `pickle`. But creating it is super quick:

```python
from dsf import mobility
import time

# how much time does it take to load the road network from file?

start_time = time.time()
rn = mobility.RoadNetwork()
edges_filepath = "../input/true_edges.csv"
rn.importEdges(edges_filepath)
rn.makeRoundabout(72)
end_time = time.time()
print(f"Time taken to load road network: {end_time - start_time} seconds")
```
